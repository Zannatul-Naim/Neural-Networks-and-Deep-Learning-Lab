```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from tqdm import tqdm

torch.manual_seed(42)
np.random.seed(42)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

train_dataset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
test_dataset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=256, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=256, shuffle=False)

classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')

class Autoencoder(nn.Module):
    def __init__(self, latent_dim=2):
        super(Autoencoder, self).__init__()
        
        # Encoder
        self.encoder = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, stride=2, padding=1),  # 16x16
            nn.ReLU(),
            nn.Conv2d(32, 64, kernel_size=3, stride=2, padding=1),  # 8x8
            nn.ReLU(),
            nn.Conv2d(64, 128, kernel_size=3, stride=2, padding=1),  # 4x4
            nn.ReLU(),
            nn.Flatten(),
            nn.Linear(128*4*4, 256),
            nn.ReLU(),
            nn.Linear(256, latent_dim)
        )
        
        # Decoder
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, 256),
            nn.ReLU(),
            nn.Linear(256, 128*4*4),
            nn.ReLU(),
            nn.Unflatten(1, (128, 4, 4)),
            nn.ConvTranspose2d(128, 64, kernel_size=3, stride=2, padding=1, output_padding=1),  # 8x8
            nn.ReLU(),
            nn.ConvTranspose2d(64, 32, kernel_size=3, stride=2, padding=1, output_padding=1),  # 16x16
            nn.ReLU(),
            nn.ConvTranspose2d(32, 3, kernel_size=3, stride=2, padding=1, output_padding=1),  # 32x32
            nn.Tanh()
        )
    
    def forward(self, x):
        latent = self.encoder(x)
        reconstructed = self.decoder(latent)
        return reconstructed, latent

latent_dim = 2  # 2D features for visualization
autoencoder = Autoencoder(latent_dim).to(device)

criterion = nn.MSELoss()
optimizer = optim.Adam(autoencoder.parameters(), lr=0.001)

num_epochs = 100
for epoch in range(num_epochs):
    autoencoder.train()
    total_loss = 0
    
    for images, _ in tqdm(train_loader, desc=f'Epoch {epoch+1}/{num_epochs}'):
        images = images.to(device)
        
        # Forward pass
        reconstructed, _ = autoencoder(images)
        loss = criterion(reconstructed, images)
        
        # Backward pass and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
    
    avg_loss = total_loss / len(train_loader)
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.4f}')

## 4. Extract features using the trained autoencoder
def extract_autoencoder_features(loader, model):
    model.eval()
    features = []
    labels = []
    
    with torch.no_grad():
        for images, targets in loader:
            images = images.to(device)
            _, latent = model(images)
            features.append(latent.cpu())
            labels.append(targets)
    
    return torch.cat(features, dim=0).numpy(), torch.cat(labels, dim=0).numpy()

# Get autoencoder features
autoencoder_features, autoencoder_labels = extract_autoencoder_features(test_loader, autoencoder)

## 5. Extract features using pre-trained VGG16
def get_vgg16_features():
    # Load pre-trained VGG16 without top classification layers
    vgg16 = torchvision.models.vgg16(pretrained=True)
    vgg16.classifier = vgg16.classifier[:4]  
    vgg16 = vgg16.to(device)
    vgg16.eval()
    
    features = []
    labels = []
    
    with torch.no_grad():
        for images, targets in test_loader:
            images = images.to(device)
            output = vgg16(images)
            features.append(output.cpu())
            labels.append(targets)
    
    return torch.cat(features, dim=0).numpy(), torch.cat(labels, dim=0).numpy()

vgg16_features, vgg16_labels = get_vgg16_features()

def reduce_dimensions(features, method='pca', n_components=2):
    if method == 'pca':
        reducer = PCA(n_components=n_components)
    elif method == 'tsne':
        reducer = TSNE(n_components=n_components, perplexity=30, n_iter=300)
    else:
        raise ValueError("Method must be 'pca' or 'tsne'")
    
    return reducer.fit_transform(features)

# Reduce VGG16 features
vgg16_pca = reduce_dimensions(vgg16_features, 'pca')
vgg16_tsne = reduce_dimensions(vgg16_features, 'tsne')

## 7. Visualization
def plot_features(features, labels, title, ax):
    scatter = ax.scatter(features[:, 0], features[:, 1], c=labels, cmap='tab10', alpha=0.6)
    ax.set_title(title)
    ax.legend(handles=scatter.legend_elements()[0], labels=classes)

fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Autoencoder features (already 2D)
plot_features(autoencoder_features, autoencoder_labels, 'Autoencoder 2D Features', axes[0, 0])

# VGG16 with PCA
plot_features(vgg16_pca, vgg16_labels, 'VGG16 Features (PCA Reduced)', axes[0, 1])

# VGG16 with t-SNE
plot_features(vgg16_tsne, vgg16_labels, 'VGG16 Features (t-SNE Reduced)', axes[1, 0])

autoencoder_norm = (autoencoder_features - autoencoder_features.mean(0)) / autoencoder_features.std(0)
vgg16_pca_norm = (vgg16_pca - vgg16_pca.mean(0)) / vgg16_pca.std(0)

axes[1, 1].scatter(autoencoder_norm[:, 0], autoencoder_norm[:, 1], c='blue', alpha=0.3, label='Autoencoder')
axes[1, 1].scatter(vgg16_pca_norm[:, 0], vgg16_pca_norm[:, 1], c='red', alpha=0.3, label='VGG16+PCA')
axes[1, 1].set_title('Normalized Feature Space Comparison')
axes[1, 1].legend()

plt.tight_layout()
plt.show()
```

```python
def visualize_reconstructions(loader, autoencoder, vgg_model, num_images=8):
    data_iter = iter(loader)
    images, labels = next(data_iter)
    images = images.to(device)
    
    with torch.no_grad():
        ae_recon, _ = autoencoder(images) 
        ae_recon = ae_recon 
        
        # VGG16 features
        vgg_features = vgg_model(images)
    
    def denormalize(tensor):
        tensor = tensor.cpu().numpy()
        tensor = tensor.transpose(0, 2, 3, 1)  # Change from CxHxW to HxWxC
        tensor = tensor * 0.5 + 0.5  # Scale from [-1,1] to [0,1]
        return np.clip(tensor, 0, 1)
    
    original_images = denormalize(images)
    ae_images = denormalize(ae_recon)
    
    vgg_feats = vgg_features.cpu().numpy()
    vgg_feats = vgg_feats[:, :3]  
    
    vgg_feats = (vgg_feats - vgg_feats.min(axis=0)) / (vgg_feats.max(axis=0) - vgg_feats.min(axis=0))
    vgg_feats = vgg_feats.reshape(-1, 1, 1, 3)  
    
    plt.figure(figsize=(15, 8))
    
    for i in range(min(num_images, images.shape[0])):
        plt.subplot(3, num_images, i+1)
        plt.imshow(original_images[i])
        if i == 0:
            plt.title('Original\n' + classes[labels[i].item()], fontsize=10)
        else:
            plt.title(classes[labels[i].item()], fontsize=10)
        plt.axis('off')
        
        plt.subplot(3, num_images, i+1+num_images)
        plt.imshow(ae_images[i])
        if i == 0:
            plt.title('Autoencoder\nReconstruction', fontsize=10)
        plt.axis('off')
        
        plt.subplot(3, num_images, i+1+2*num_images)
        plt.imshow(vgg_feats[i])
        if i == 0:
            plt.title('VGG16\nFeature Map', fontsize=10)
        plt.axis('off')
    
    plt.tight_layout()
    plt.subplots_adjust(top=0.85)
    plt.suptitle('Comparison between AE reconstruction and VGG16 feature map', fontsize=14)
    plt.show()

vgg16 = torchvision.models.vgg16(pretrained=True)
vgg16.classifier = vgg16.classifier[:4] 
vgg16 = vgg16.to(device)
vgg16.eval()

# Visualize reconstructions
visualize_reconstructions(test_loader, autoencoder, vgg16)
```
